```python
import tabu
import random
import datetime
city = []
#eg=
for line in open("tsp.txt"):
    place,lon,lat = line.strip().split(" ")
    city.extend([(place,(lon,lat))])

city1=city.copy()
city_list=[]
cc=city1.copy()
city_list.append(cc)
i=0
while i < 28:
    city1.append(city1[0])
    city1.remove(city1[0])
    cc=city1.copy()
    city_list.append(cc)
    i=i+1
```

```python
from multiprocessing import Pool
if __name__ ==  '__main__': 
 num_processors = 3
 p=Pool(processes = num_processors)
 output = p.map(tabu.parallelfunc,city_list)
output
```

```python
def finalanswer(output):
    distances=[]
    for i in output:
        distances.append(i[0])
    return(distances.index(min(distances)),output[distances.index(min(distances))][0],output[distances.index(min(distances))][1])
```

```python
answer1=finalanswer(output)
answer1
```
